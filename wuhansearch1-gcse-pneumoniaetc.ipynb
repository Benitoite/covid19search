```python
import requests
import pandas as pd
import sys, json
import json
import string
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
from config import csekey
import datetime as dt
import time as t
```

```python
start_date = dt.date(2018, 1, 1)
end_date = dt.date(2021, 11, 3)
delta = dt.timedelta(days=1)
results = []
url = 'https://www.googleapis.com/customsearch/v1/siterestrict'

while start_date <= end_date:

    # Construct the sortby date
    date = str(start_date.year) + start_date.strftime('%m') + start_date.strftime('%d')
    params = { 'sort': 'date:r:'+date+':'+date, 'q': 'pneumonia+OR+cough+OR+diarrhea', 'cx': 'e22950bd22e534ff4', 'key': csekey}

    # Poll the API service
    response = requests.get(url, params=params).json()
    hits=response['searchInformation']['totalResults']
    results.append([start_date,hits])
    print("Got " + date + " with " + str(hits) + " hits.")
    
    # Increment the sortby date
    start_date += delta

    # Throttle for a second
    t.sleep(1)

```

```python
df = pd.DataFrame(results, columns=["date","totalResults"])

# Decode vaules above 10
for i,row in df.iterrows():
    if int(row[1]) > 100:
        df.iloc[i][1] = int(int(row[1])//1000-156)

# Drop 0 result rows
for i,row in df.iterrows():
    if int(row[1]) == 0:
        df.drop(i,inplace=True)
```
